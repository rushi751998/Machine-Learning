```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Parse_dates converts in to date format and dayfirst= True asignes it to firsrt

```python
df = pd.read_csv('https://raw.githubusercontent.com/rushi751998/Practice-/main/machine%20learning/covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# keeping only requird columns
df = df[['Date', 'State/UnionTerritory', 'Cured',
       'Deaths', 'Confirmed']]



# renaming columns name
df.columns = ['Date', 'State', 'Cured',
       'Deaths', 'Confirmed']

# making columns name in lower case
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
# sorting data w.r.t. conformend cases
max_conformed_case = today.sort_values(by='confirmed',ascending=False)
```
